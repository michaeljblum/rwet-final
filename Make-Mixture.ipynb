```python
import io

with io.open('beckett-fake-ish.txt', encoding='utf8', errors='ignore') as l:
    fake = l.read()
    l.close()
    
with io.open('beckett.txt', encoding='utf8', errors='ignore') as m:
    real = m.read()
    m.close()
    
with io.open('beckettpoems.txt', encoding='utf8', errors='ignore') as m:
    poems = m.read()
    m.close()

with io.open('output.txt', encoding='utf8', errors='ignore') as n:
    sports = n.read()
    n.close()
```

```python
from nltk import sent_tokenize
```

```python
real_sent = sent_tokenize(real)
```

```python
# print(sports)
```

```python
# split sports on new lines and sentences, then filter out all sentences in the sports_sentences that
# are larger than X characters...
import re

sports_sent = re.split('\.|\n', sports)

sports_chunk = []
for x in sports_sent:
    if len(x) < 80 and len(x) > 5:
        x = re.sub(r'[^\w\s]', '', x)
        x = re.sub(r'[0-9]+', '', x)
        sports_chunk.append(x)
    
sports_chunk = list(filter(None, sports_chunk))

print(sports_chunk)
```

```python
fake = fake.split("\n")
fake = list(filter(None, fake))
# for x in fake:
#     x = re.sub('[€]', '', x)

    
print(fake)
```

```python
poems = poems.split("\n")
poems = list(filter(None, poems))
print(poems)
```

```python
def mix(fake, sports_chunk, poems, real_sent):
    mixed = []
    i = 0
    while i < len(poems):
        mixed.append(fake[i])
        mixed.append(sports_chunk[i])
        mixed.append(real_sent[i])
        mixed.append(poems[i])
        i += 1
    mixed = "\n".join(mixed)
    with io.open('mixture.txt', 'xt', encoding='utf8', errors='ignore') as f:
        # Write to the file
        f.write(mixed)
        # Close the connection to the file
        f.close()
    return mixed
```

```python
mix(fake, sports_chunk, poems, real_sent)
```
