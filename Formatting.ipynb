```python
import io
from textwrap3 import wrap

def formatting(input):
    with io.open(input, 'rt', encoding='utf8', errors='ignore') as f:
        raw = f.read()
        f.close()
#     wrapped = ''
    x = wrap(raw, 40)
    for i in range(len(x)):
        print(x[i].center(100))

formatting("output/v2-101-best.txt.txt")

# string_centered = string.center(79, '.')
# string_centered
```
